```python
import pandas as pd
s = pd.read_csv("insurance.csv")
s.info()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,4, figsize = (20,4))
ax[0].hist(s['age'], edgecolor = 'black', color = 'green',bins = 15,label = 'bins = 15') 
ax[0].legend() 
ax[0].set_title("Гистограмма возраста")
ax[1].hist(s['bmi'], edgecolor = 'black', color = 'green',bins = 15,label = 'bins = 15') 
ax[1].legend() 
ax[1].set_title("Гистограмма индекса массы тела")
ax[2].hist(s['children'], edgecolor = 'black', color = 'green',bins = 5,label = 'bins = 5') 
ax[2].legend() 
ax[2].set_title("Гистограмма детей")
ax[3].hist(s['charges'], edgecolor = 'black', color = 'green',bins = 7,label = 'bins = 7') 
ax[3].legend() 
ax[3].set_title("Гистограмма расходов")
plt.show()
```

```python
import numpy as np
import scipy.stats as sts

meanBM = np.mean(s['bmi'])
modaBM = np.mean(sts.mode(s['bmi']).mode)
medBM = np.median(s['bmi'])
meanCH = np.mean(s['charges'])
modaCH = np.mean(sts.mode(s['charges']).mode)
medCH = np.median(s['charges'])

fig, ax = plt.subplots(1,2, figsize = (12,6))
ax[0].hist(s['bmi'], edgecolor = 'black', color = 'green',bins = 15,label = 'bins = 15') 
ax[0].vlines(meanBM,0,len(s['bmi'])/6, label='Mean = '+str("{:.2f}".format(meanBM)), color='red')
ax[0].vlines(modaBM,0,len(s['bmi'])/6, label='Moda = '+str("{:.2f}".format(modaBM)), color='blue')
ax[0].vlines(medBM,0,len(s['bmi'])/6, label='Med = '+str("{:.2f}".format(medBM)), color='yellow')
ax[0].legend() 
ax[0].set_title("Гистограмма индекса массы тела")
ax[1].hist(s['charges'], edgecolor = 'black', color = 'green',bins = 7,label = 'bins = 7') 
ax[1].vlines(meanCH,0,len(s['charges'])/1.75, label='Mean = '+str("{:.2f}".format(meanCH)), color='red')
ax[1].vlines(modaCH,0,len(s['charges'])/1.75, label='Moda = '+str("{:.2f}".format(modaCH)), color='blue')
ax[1].vlines(medCH,0,len(s['charges'])/1.75, label='Med = '+str("{:.2f}".format(medCH)), color='yellow')
ax[1].legend() 
ax[1].set_title("Гистограмма расходов")
plt.show()
```

```python
stdBM = s['bmi'].std()
razBM = s['bmi'].max() - s['bmi'].min()
iqrBM = sts.iqr(s['bmi'], interpolation = 'midpoint')

stdCH = s['charges'].std()
razCH = s['charges'].max() - s['charges'].min()
iqrCH = sts.iqr(s['charges'], interpolation = 'midpoint')

fig, ax = plt.subplots(1,2, figsize = (12,6))
ax[0].hist(s['bmi'], edgecolor = 'black', color = 'green',bins = 15,label = 'bins = 15') 
ax[0].vlines(stdBM,0,len(s['bmi'])/6, label='Std = '+str("{:.2f}".format(stdBM)), color='red')
ax[0].vlines(razBM,0,len(s['bmi'])/6, label='Размах = '+str("{:.2f}".format(razBM)), color='blue')
ax[0].vlines(iqrBM,0,len(s['bmi'])/6, label='IQR = '+str("{:.2f}".format(iqrBM)), color='yellow')
ax[0].legend() 
ax[0].set_title("Гистограмма индекса массы тела")
ax[1].hist(s['charges'], edgecolor = 'black', color = 'green',bins = 7,label = 'bins = 7') 
ax[1].vlines(stdCH,0,len(s['charges'])/1.75, label='Std = '+str("{:.2f}".format(stdCH)), color='red')
ax[1].vlines(razCH,0,len(s['charges'])/1.75, label='Размах = '+str("{:.2f}".format(razCH)), color='blue')
ax[1].vlines(iqrCH,0,len(s['charges'])/1.75, label='IQR = '+str("{:.2f}".format(iqrCH)), color='yellow')
ax[1].legend() 
ax[1].set_title("Гистограмма расходов")
plt.show()
```

```python
import random

means50 = []
for i in range(300):
    meanNow = []
    n1 = 2
    for j in range(int(len(s['bmi'])/n1)):
        meanNow.append(s['bmi'][2*i+j])
    means50.append(np.mean(meanNow))

means350 = []
for i in range(300):
    meanNow = []
    n1 = 3
    for j in range(int(len(s['bmi'])/n1)):
        meanNow.append(s['bmi'][2*i+j])
    means350.append(np.mean(meanNow))

means500 = []
for i in range(300):
    meanNow = []
    n1 = 4
    for j in range(int(len(s['bmi'])/n1)):
        meanNow.append(s['bmi'][2*i+j])
    means500.append(np.mean(meanNow))


mean50 = np.mean(means50)
sd50 = np.std(means50)    

mean350 = np.mean(means350)
sd350 = np.std(means350)    

mean500 = np.mean(means500)
sd500 = np.std(means500)    
    
print('Среднее арифмитическое для выборки 650: ', mean50)
print('Стандатное отклонение для выборки 650: ', sd50)
print('Среднее арифмитическое для выборки 430: ', mean350)
print('Стандатное отклонение для выборки 430: ', sd350)
print('Среднее арифмитическое для выборки 320: ', mean500)
print('Стандатное отклонение для выборки 320: ', sd500)

means50.sort()
means350.sort()
means500.sort()


pdf50 = sts.norm.pdf(means50,mean50,sd50)
pdf50 = pdf50*100*sd50

pdf350 = sts.norm.pdf(means350,mean350,sd350)
pdf350 = pdf350*100*sd350

pdf500 = sts.norm.pdf(means500,mean500,sd500)
pdf500 = pdf500*100*sd500


fig, ax = plt.subplots(1,3, figsize=(15,5))

ax[0].plot(means50,pdf50,color='crimson')
ax[0].hist(means50, edgecolor = 'black', color = 'green',bins = 15)
ax[0].set_title('Распределение выборок')
ax[0].set_xlabel("Количество")
ax[0].set_ylabel("Средняя величина")

ax[1].plot(means350,pdf350,color='crimson')
ax[1].hist(means350, edgecolor = 'black', color = 'green',bins = 15)
ax[1].set_title('Распределение выборок')
ax[1].set_xlabel("Количество")
ax[1].set_ylabel("Средняя величина")

ax[2].plot(means500,pdf500,color='crimson')
ax[2].hist(means500, edgecolor = 'black', color = 'green',bins = 15)
ax[2].set_title('Распределение выборок')
ax[2].set_xlabel("Количество")
ax[2].set_ylabel("Средняя величина")

plt.show()
```
