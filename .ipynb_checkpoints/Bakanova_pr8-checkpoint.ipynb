```python
import numpy as np

street = [80, 98, 75, 91, 78]
garage = [100, 82, 105, 89, 102]
np.corrcoef(street, garage)[0,1]
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,7))
plt.grid(True)
plt.title('Диаграмма рассеяния', fontsize=25)
plt.xlabel('Число машин на улице')
plt.ylabel('Число машин в гараже')
plt.scatter(street,garage,marker='o',color='crimson')
```

```python
import pandas as pd
df = pd.read_csv("bitcoin.csv")
df.info()
```

```python
df_14=df[0:len(df)-14]
```

```python
import numpy as np

from sklearn.linear_model import LinearRegression

x = df_14['Unnamed: 0']
y = df_14['close']

n = np.size(x)

x = np.array(x,type(float))
y = np.array(y,type(float))

x=x[1800:len(x)]
y=y[1800:len(y)]

regression = LinearRegression()
regression.fit(x.reshape(-1,1),y)


fig2 = plt.figure(figsize=(9,6))

plt.scatter(x,y, color="blue", marker="o", s = 30)


plt.plot(x, regression.predict(x.reshape(-1,1)), color = "g") #построение линии регресии
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
