```python
import scipy

bm = pd.read_csv("bmi.csv")

nor=pd.Series(bm['bmi'].loc[bm['region']=='northwest'])
sou=pd.Series(bm['bmi'].loc[bm['region']=='southwest'])

s1 = np.random.choice(nor, 40)
s2 = np.random.choice(sou, 40)

test1 = scipy.stats.shapiro(s1)
test2 = scipy.stats.shapiro(s2)

print(test1,'\n',test2)

test = scipy.stats.bartlett(s1, s2)

print(test)

Stud = scipy.stats.ttest_ind(s1,s2)
print(Stud)
```
