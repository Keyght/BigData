```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
from sklearn import preprocessing
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import time

start_time = time.time()

mnist = load_digits()
s = pd.DataFrame(mnist['data'])

scaler = preprocessing.MinMaxScaler()
s = pd.DataFrame(scaler.fit_transform(s))
T=TSNE(n_components=2, perplexity=5, random_state=1797)
TSNE_features = T.fit_transform(s)
print("--- %s seconds ---" % (time.time() - start_time))

DATA=s.copy()
DATA['x']=TSNE_features[:,0]
DATA['y']=TSNE_features[:,1]
fig = plt.figure(figsize=(10,10))
sns.scatterplot(x='x',y='y',hue=mnist['target'],data=DATA, palette='bright')
plt.show()
```
